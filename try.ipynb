```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tweets_df = pd.DataFrame({
    "tweet": ["Hello", None, "How are you?", "Great", None],
    "likes": [10, 5, None, 8, None]
})

# Heatmap of missing data
sns.heatmap(tweets_df.isnull(), yticklabels=False, cbar=True, cmap="Blues")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

reviews = [
    "I love this product, it is amazing!",
    "Terrible experience, will not buy again.",
    "Not bad, but could be better.",
    "Excellent quality and fast shipping."
]

vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(reviews)

print("Shape:", bow.shape)
print("Vocabulary:", vectorizer.get_feature_names_out())
print("Feature matrix:\n", bow.toarray())


```
